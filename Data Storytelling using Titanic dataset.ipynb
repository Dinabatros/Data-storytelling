#  Data Storytelling using Titanic dataset

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
df.head(50)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df.isnull().sum()
```

# Question 1: What sorts of people were more likely to survive?

Let us find out what sorts of people had more chances to survive and to do this, we are going to explore the survival rate by different factors such as class, gender, number of siblings/spouses, number of parents/children, and port of embarkation.

```python
def survival_rate(factor):
    sns.barplot(x=factor, y='Survived', data=df)
    plt.title(f'Survival Rate by {factor}')
    plt.show()
    
factors = ['SibSp', 'Parch', 'Embarked']

for factor in factors:
    survival_rate(factor)
```

Passengers with fewer siblings/spouses (SibSp) and parents/children (Parch) had slightly higher survival rates.
Those embarking from Cherbourg (C) had a higher survival rate compared to those from Queenstown (Q) and Southampton (S).

# Question 2: Did passengers in first class have higher survival rates compared to those in second and third class?

```python
# Survival rate by class
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Class')
plt.show()
```

Passengers in first class (Pclass=1) had a significantly higher survival rate compared to second (Pclass=2) and third class (Pclass=3). This aligns with the historical precedence of prioritizing higher-class passengers during the evacuation.

# Question 3: What was the age range of survivors, and which age group had the highest survival rate?

```python
# Age distirbution
sns.histplot(df[df['Survived'] == 1]['Age'], bins=20, kde=True)
plt.title('Age Distribution of Survivors')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Survival rate by age group
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 35, 60, 80], labels=['Child', 'Teen', 'Adult', 'Middle Age', 'Senior'])
sns.barplot(x='AgeGroup', y='Survived', data=df)
plt.title('Survival Rate by Age Group')
plt.show()
```

The age distribution of survivors skewed towards younger ages, with a notable peak in survival rates among children (AgeGroup='Child').
Adults (AgeGroup='Adult') also had a relatively high survival rate, while seniors (AgeGroup='Senior') had the lowest survival rate among the age groups analyzed.

As we can see, we have a very high survival rates of children compared to adults and seniors

# Question 4: How did gender influence survival rates on the Titanic?


```python
# Survival rate by gender
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()
```

Females had a significantly higher survival rate compared to males. This reflects the "women and children first" policy implemented during the Titanic disaster, where priority was given to women and children during evacuation.

#  Overall Conclusion:

Passengers who were female, in first class, had fewer family members aboard, embarked from Cherbourg, and were younger (especially children) had higher chances of survival.

Policy Influence: The survival patterns observed align closely with the evacuation policies practiced during the Titanic disaster, highlighting the impact of socio-economic status (class), age, and gender on survival outcomes.
